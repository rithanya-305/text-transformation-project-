```python
# 🛠️ STEP 1: Install Required Libraries
!pip install transformers datasets sentencepiece --quiet

# 🧠 STEP 2: Import Libraries
from transformers import pipeline

# 🧠 STEP 3: Load Pretrained T5 Model for Text-to-Text Generation
t2t = pipeline("text2text-generation", model="t5-base")

# ✏️ STEP 4: Sample Input Texts for Each Transformation
sample_text = "Artificial intelligence is revolutionizing industries and transforming how people interact with technology."
long_text = """
Artificial Intelligence (AI) is a field of computer science focused on building machines that can mimic human intelligence.
It encompasses machine learning, natural language processing, and robotics. AI is now being applied in healthcare, finance,
transportation, and many other sectors to improve efficiency and decision-making.
"""
incorrect_text = "He go to school every day."
fact_text = "The capital of France is Paris."

# ✅ STEP 5: Run Text Transformations

# 🔁 1. Paraphrasing
paraphrased = t2t("paraphrase: " + sample_text, max_length=100)
print("🔁 Paraphrased Text:\n", paraphrased[0]['generated_text'])

# 🧠 2. Summarization
summary = t2t("summarize: " + long_text, max_length=50)
print("\n🧠 Summary:\n", summary[0]['generated_text'])

# 🌍 3. Translation (English to French)
translation = t2t("translate English to French: " + sample_text)
print("\n🌍 Translated to French:\n", translation[0]['generated_text'])

# ✅ 4. Grammar Correction
corrected = t2t("grammar correction: " + incorrect_text)
print("\n✅ Corrected Sentence:\n", corrected[0]['generated_text'])

# ❓ 5. Question Generation
question = t2t("generate question: " + fact_text)
print("\n❓ Generated Question:\n", question[0]['generated_text'])

# 🎯 STEP 6: User-Defined Input Example (Paraphrasing)
print("\n🎯 User Example:")
user_input = "Machine learning helps computers improve automatically through experience."
user_output = t2t("paraphrase: " + user_input)
print("Original:\n", user_input)
print("Paraphrased:\n", user_output[0]['generated_text'])
```
